```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.covariance import EllipticEnvelope
import matplotlib.pyplot as plt
```

```python
# Create a sample dataset
data = {
    'Feature1': [1, 2, 3, 10, 11, 12, 20, 25, 30],
    'Feature2': [5, 6, 7, 8, 9, 15, 18, 22, 28]
}

df = pd.DataFrame(data)
```

```python
# Visualize the original data
plt.scatter(df['Feature1'], df['Feature2'])
plt.title('Original Data')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.show()

```

```python
# Use K-Means clustering to identify clusters
kmeans = KMeans(n_clusters=2)
df['Cluster'] = kmeans.fit_predict(df[['Feature1', 'Feature2']])
```

```python
# Visualize the clustered data
plt.scatter(df['Feature1'], df['Feature2'], c=df['Cluster'], cmap='viridis')
plt.title('Clustered Data')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.show()
```

```python
# Identify outliers using Elliptic Envelope
envelope = EllipticEnvelope(contamination=0.1)  # Adjust contamination based on your dataset
df['Outlier'] = envelope.fit_predict(df[['Feature1', 'Feature2']])
```

```python
# Remove outliers
df_filtered = df[df['Outlier'] != -1].drop(['Cluster', 'Outlier'], axis=1)
```

```python
# Visualize the data after removing outliers
plt.scatter(df_filtered['Feature1'], df_filtered['Feature2'])
plt.title('Data after Outlier Removal')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.show()
```
